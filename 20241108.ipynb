```python
import csv
import numpy as np
import pandas as pd
```

```python
# csv파일 불러오기

def file_open_by_numpy():
    # np.loadtxt(구분자 = ',', 데이터 타입: string)
    np_arr = np.loadtxt('test_data.CSV', delimiter=",", encoding='cp949', dtype=str)
    return np_arr
```

```python
arr = file_open_by_numpy()
print(arr)
```

```python
type(arr)
```

```python
# 데이터프레임으로 변환

df = pd.DataFrame(arr)
df
```

```python
# 컬럼명 지정하기
columns=arr[0]
arr = np.delete(arr, 0, 0)
df = pd.DataFrame(arr, columns=columns)
df
```

```python
# isna().sum() 함수를 통해 null값 개수 찾기(결측값이면 True 반환, 정상값이면 False반환)

df.isna().sum()
```

```python
# 결측값을 NULL문자열로 치환
df.replace('', 'NULL', inplace=True)
```
